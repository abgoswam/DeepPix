## Original Image (before pre-processing)

```python
import numpy as np
from scipy.misc import imread, imresize, imsave
import matplotlib.pyplot as plt

img = imread('img_orig.jpg')
print(img.dtype, img.shape)  
```

```python
plt.imshow(img)
plt.show()
```

```python
# Resize.
img_32_32 = imresize(img, (32, 32))

print(img_32_32.dtype, img_32_32.shape) 
imsave('img_32_32.jpg', img_32_32)
```
